```python
# ✅ 1. Upload CSV File
from google.colab import files
uploaded = files.upload()

import pandas as pd

# Replace 'your_file.csv' with your uploaded filename
filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

# ✅ 2. View Dataset
print("First few rows of the dataset:")
print(df.head())

# ✅ 3. Preprocessing
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Check required columns
if 'query' not in df.columns or 'intent' not in df.columns:
    raise ValueError("CSV must have 'query' and 'intent' columns")

X = df['query']
y = df['intent']

vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X)

# ✅ 4. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_vec, y, test_size=0.2, random_state=42)

# ✅ 5. Model Training
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)

# ✅ 6. Evaluation
from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# ✅ 7. Chatbot Function
def chatbot_response(user_input):
    user_vec = vectorizer.transform([user_input])
    predicted_intent = model.predict(user_vec)[0]

    response_map = {
        "greeting": "Hello! How can I help you today?",
        "order_status": "Can you share your order ID?",
        "return": "You can return a product within 30 days.",
        "complaint": "Sorry to hear that! We'll fix it.",
        "product_query": "What would you like to know about the product?",
        "thanks": "You're welcome!",
        "goodbye": "Goodbye! Have a great day!",
        # Add more intent-response pairs as needed
    }

    return response_map.get(predicted_intent, "Sorry, I didn’t understand that.")

# ✅ 8. Try Chatbot
user_input = input("Ask a question: ")
print("Bot:", chatbot_response(user_input))
```
